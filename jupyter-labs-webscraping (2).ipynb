````python
import sys

import requests
from bs4 import BeautifulSoup
import re
import unicodedata
import pandas as pd
````

and we will provide some helper functions for you to process web scraped HTML table


````python
def date_time(table_cells):
    """
    This function returns the data and time from the HTML  table cell
    Input: the  element of a table data cell extracts extra row
    """
    return [data_time.strip() for data_time in list(table_cells.strings)][0:2]

def booster_version(table_cells):
    """
    This function returns the booster version from the HTML  table cell 
    Input: the  element of a table data cell extracts extra row
    """
    out=''.join([booster_version for i,booster_version in enumerate( table_cells.strings) if i%2==0][0:-1])
    return out

def landing_status(table_cells):
    """
    This function returns the landing status from the HTML table cell 
    Input: the  element of a table data cell extracts extra row
    """
    out=[i for i in table_cells.strings][0]
    return out


def get_mass(table_cells):
    mass=unicodedata.normalize("NFKD", table_cells.text).strip()
    if mass:
        mass.find("kg")
        new_mass=mass[0:mass.find("kg")+2]
    else:
        new_mass=0
    return new_mass


def extract_column_from_header(row):
    """
    This function returns the landing status from the HTML table cell 
    Input: the  element of a table data cell extracts extra row
    """
    if (row.br):
        row.br.extract()
    if row.a:
        row.a.extract()
    if row.sup:
        row.sup.extract()
        
    colunm_name = ' '.join(row.contents)
    
    # Filter the digit and empty names
    if not(colunm_name.strip().isdigit()):
        colunm_name = colunm_name.strip()
        return colunm_name    
````

Next, we just need to iterate through the `<th>` elements and apply the provided `extract_column_from_header()` to extract column name one by one


````python
column_names = []

# Apply find_all() function with `th` element on first_launch_table
# Iterate each th element and apply the provided extract_column_from_header() to get a column name
# Append the Non-empty column name (`if name is not None and len(name) > 0`) into a list called column_names
import re

# Usa la función del notebook si existe; si no, define un fallback
try:
    extract_column_from_header
except NameError:
    def extract_column_from_header(th):
        """
        Extrae el texto de un <th>, limpia saltos de línea, <br>, y referencias [a], [1], etc.
        Devuelve un nombre de columna legible.
        """
        # texto plano con separadores donde hay <br> u otras etiquetas
        text = th.get_text(separator=' ', strip=True)
        # elimina notas/referencias entre corchetes
        text = re.sub(r'\[[^\]]*\]', '', text)
        # normaliza espacios
        text = re.sub(r'\s+', ' ', text).strip()
        return text or None

# 3) Recorrer los th y construir la lista de nombres
column_names = []
for th in first_launch_table.find_all('th'):
    name = extract_column_from_header(th)
    if name:  # evitar vacíos
        column_names.append(name)

# Ver resultado
print(column_names)
````

````python
import re
import pandas as pd

# -------------------------------------------------------
# 0) Fallbacks (solo si tu notebook NO las definió antes)
# -------------------------------------------------------
try:
    date_time
except NameError:
    def date_time(td):
        """Devuelve [date_str, time_str] buscando patrones comunes en la celda."""
        text = td.get_text(" ", strip=True)
        m = re.search(r'(\d{1,2}\s+[A-Za-z]+\s+\d{4})[, ]+(\d{2}:\d{2}(?::\d{2})?)', text)
        if m:
            return [m.group(1), m.group(2)]
        # Último recurso: separar por 'UTC'
        parts = text.replace('(UTC)', '').split(',')
        date = parts[0].strip() if parts else text
        time = parts[1].strip() if len(parts) > 1 else None
        return [date, time]

try:
    booster_version
except NameError:
    def booster_version(td):
        """Extrae la versión del booster (texto limpio)."""
        text = td.get_text(" ", strip=True)
        text = re.sub(r'\[[^\]]*\]', '', text)  # quita referencias [a], [1]...
        text = re.sub(r'\s+', ' ', text).strip()
        return text or None

try:
    get_mass
except NameError:
    def get_mass(td):
        """Devuelve la masa del payload (en kg como número) si aparece, si no, texto limpio."""
        text = td.get_text(" ", strip=True)
        # Busca 'x,xxx kg'
        m = re.search(r'([\d,]+)\s*kg', text, re.I)
        if m:
            return float(m.group(1).replace(',', ''))
        return None  # o text si prefieres dejarlo como string

try:
    landing_status
except NameError:
    def landing_status(td):
        """Devuelve el estado de aterrizaje; toma el primer string 'limpio'."""
        s = " ".join(list(td.stripped_strings))
        s = re.sub(r'\[[^\]]*\]', '', s).strip()
        return s or None

# -------------------------------------------------------
# 1) Crear launch_dict a partir de column_names y limpiar
# -------------------------------------------------------
launch_dict = dict.fromkeys(column_names, None)

# Quita la columna de fecha-hora si viene con nombre "ruidoso"
for k in list(launch_dict.keys()):
    if k and k.lower().startswith("date and time"):
        del launch_dict[k]

# Inicializa las listas que necesitamos (coinciden con el enunciado)
launch_dict['Flight No.']      = []
launch_dict['Launch site']     = []
launch_dict['Payload']         = []
launch_dict['Payload mass']    = []
launch_dict['Orbit']           = []
launch_dict['Customer']        = []
launch_dict['Launch outcome']  = []
# Columnas nuevas
launch_dict['Version Booster'] = []
launch_dict['Booster landing'] = []
launch_dict['Date']            = []
launch_dict['Time']            = []

# -------------------------------------------------------
# 2) Recorrer las tablas de lanzamientos y extraer filas
# -------------------------------------------------------
extracted_row = 0
tables = soup.find_all('table', "wikitable plainrowheaders collapsible")

for table_number, table in enumerate(tables):
    for rows in table.find_all("tr"):
        # ¿La cabecera de fila (th) contiene el número de vuelo?
        flag = False
        if rows.th and rows.th.string:
            flight_number = rows.th.string.strip()
            flag = flight_number.isdigit()

        row = rows.find_all('td')
        if flag and row:
            extracted_row += 1

            # --- Flight No. ---
            launch_dict['Flight No.'].append(int(flight_number))

            # --- Date & Time ---
            dt = date_time(row[0])
            date = dt[0].strip(',') if dt and len(dt) > 0 else None
            time = dt[1] if dt and len(dt) > 1 else None
            launch_dict['Date'].append(date)
            launch_dict['Time'].append(time)

            # --- Version Booster ---
            bv = booster_version(row[1])
            if not bv and row[1].a and row[1].a.string:
                bv = row[1].a.string
            launch_dict['Version Booster'].append(bv)

            # --- Launch site ---
            launch_site = row[2].a.string if (row[2].a and row[2].a.string) else row[2].get_text(" ", strip=True)
            launch_dict['Launch site'].append(launch_site)

            # --- Payload ---
            payload = row[3].a.string if (row[3].a and row[3].a.string) else row[3].get_text(" ", strip=True)
            launch_dict['Payload'].append(payload)

            # --- Payload mass ---
            payload_mass = get_mass(row[4])
            launch_dict['Payload mass'].append(payload_mass)

            # --- Orbit ---
            orbit = row[5].a.string if (row[5].a and row[5].a.string) else row[5].get_text(" ", strip=True)
            launch_dict['Orbit'].append(orbit)

            # --- Customer ---
            customer = row[6].a.string if (row[6].a and row[6].a.string) else row[6].get_text(" ", strip=True)
            launch_dict['Customer'].append(customer)

            # --- Launch outcome ---
            launch_outcome = next(iter(row[7].stripped_strings), None)
            launch_dict['Launch outcome'].append(launch_outcome)

            # --- Booster landing ---
            booster_landing = landing_status(row[8]) if len(row) > 8 else None
            launch_dict['Booster landing'].append(booster_landing)

# -------------------------------------------------------
# 3) DataFrame final y exportación
# -------------------------------------------------------
# Opción 1 (como en el enunciado, usando Series):
df = pd.DataFrame({k: pd.Series(v) for k, v in launch_dict.items()})

# Opción 2 (más simple, hace lo mismo):
# df = pd.DataFrame(launch_dict)

print(df.shape)
df.head()
````
