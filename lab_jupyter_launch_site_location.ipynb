```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
from math import radians, sin, cos, sqrt, atan2
import folium

# Si ya tienes calculate_distance(), omite esta función
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1)*cos(lat2)*sin(dlon/2)**2
    c = 2*atan2(sqrt(a), sqrt(1-a))
    return R*c  # km

def add_proximity(site_map, site_lat, site_lon, poi_lat, poi_lon, label, line_color="purple"):
    """Crea marcador de distancia al POI y dibuja PolyLine al launch site."""
    dist_km = calculate_distance(site_lat, site_lon, poi_lat, poi_lon)
    # Marcador en el POI
    folium.Marker(
        [poi_lat, poi_lon],
        tooltip=label,
        icon=folium.DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html=f"<div style='font-size:12px;color:#222;background:rgba(255,255,255,.8);"
                 f"padding:2px 4px;border-radius:4px;'>{label}: <b>{dist_km:0.2f} km</b></div>"
        )
    ).add_to(site_map)
    # Punto visual del POI
    folium.CircleMarker([poi_lat, poi_lon], radius=4, color=line_color, fill=True, fill_opacity=.9).add_to(site_map)
    # Línea POI ↔ Launch Site
    folium.PolyLine([[site_lat, site_lon],[poi_lat, poi_lon]],
                    weight=2, opacity=.8, color=line_color).add_to(site_map)
    return dist_km
```

```python
# --- POIs de ejemplo para CCAFS LC-40 (cámbialos si capturas otros con MousePosition) ---
poi_city      = ("Closest City: Cape Canaveral", 28.3922, -80.6077)   # ciudad
poi_railway   = ("Closest Railway: NASA Railroad", 28.5247, -80.6463) # vía férrea aprox
poi_highway   = ("Closest Highway: FL-401",       28.4156, -80.6323)  # autopista/arteria
poi_coast     = ("Closest Coastline",             28.5637, -80.5716)  # costa muy cercana

dist_city   = add_proximity(site_map, launch_site_lat, launch_site_lon, poi_city[1],   poi_city[2],   poi_city[0],   "darkred")
dist_rail   = add_proximity(site_map, launch_site_lat, launch_site_lon, poi_railway[1],poi_railway[2],poi_railway[0], "darkgreen")
dist_hwy    = add_proximity(site_map, launch_site_lat, launch_site_lon, poi_highway[1],poi_highway[2],poi_highway[0], "darkblue")
dist_coast  = add_proximity(site_map, launch_site_lat, launch_site_lon, poi_coast[1],  poi_coast[2],  poi_coast[0],   "orange")

# Renderiza o guarda si no puedes Trust Notebook
site_map.save("spacex_proximities.html")
from IPython.display import IFrame
IFrame("spacex_proximities.html", width="100%", height=600)
```
